```python
def get_VSM_bf_data(excel_sheet, specimen_index, frac = 0.1, frac_der = 0.1):
    Blist = excel_sheet.iloc[:,0 + specimen_index * 3].dropna().tolist()[5:] # field list, unit is mT
    Mlist = excel_sheet.iloc[:,1 + specimen_index * 3].dropna() .tolist()[5:]# moment list, unit is Am2/kg
    Blist = [-i for i in Blist]
    Mlist = [(i-min(Mlist)) for i in Mlist]   

    logBlist = np.log10(Blist)

    lowess = sm.nonparametric.lowess
    lowessed = lowess(Mlist, Blist, frac=frac, xvals=np.array(Blist))
    
#     print(Blist, Mlist)
    
    # use the lowess curved to calculate a median destructive field
    # the definition for the median destructive field is the strength of the alternating magnetic field required to eliminate one-half of the remanent saturation magnetization
    saturation_m = max(lowessed)
    median_m = 0.5*saturation_m
    MDF = -1 # initialize fake value for MDF
    for i in range(len(lowessed)-1):
        this_difference = lowessed[i]-median_m
        next_difference = lowessed[i+1]-median_m
        if this_difference == 0:
            MDF = Blist[i]
            break
        elif this_difference > 0 and next_difference < 0:
            MDF = (Blist[i] + Blist[i+1])/2
            break
            
    Der = np.diff(lowessed)/np.diff(np.log10(Blist))
    Der = [-j for j in Der]
    Der = Der[1:]
    Der = [k/max(Der) for k in Der]
#     print(len(logBlist), len(Mlist), lowessed.shape, len(Der))
    lowessed_der = lowess(Der, logBlist[2:], frac=frac_der, xvals=logBlist[2:])
    specimen_name = excel_sheet.columns[1 + specimen_index * 3]
    return {'Blist':Blist, 'Log_Blist':logBlist, 'Mlist':Mlist, 'Name':specimen_name, 'Lowessed':lowessed, 'Derivative':Der, 'Lowessed_der':lowessed_der, 'MDM': median_m, 'MDF': MDF}
```

```python
def skew_GaussFunc(x, amp, mu, skew, sigma):
    return amp * np.exp(-0.5 * ((x - mu) / sigma)**2) * (1+real_erf((skew*x-mu)/sigma/np.sqrt(2)))

def plot_results_1_component(example_data, trace_dataframe, n = 100):
    fig, ax = plt.subplots(figsize = (15,6))
    ax.scatter(example_data[0], example_data[1], color = 'grey', alpha = 0.6)
    x = np.linspace(0, example_data[0].max(), 100)
    def skew_Func(x, amp, mu, skew, sigma):
        return amp * np.exp(-0.5 * ((x - mu) / sigma)**2) * (1+real_erf((skew*x-mu)/sigma/np.sqrt(2)))
    
    for i in range(n):
        sample_row = trace_dataframe.sample(axis=0, n=1)
        y1 = skew_Func(x, amp =np.array(sample_row['amp']),
                      mu = np.array(sample_row['mu']), 
                      skew = np.array(sample_row['skew']),
                      sigma = np.array(sample_row['sigma']))
        y = y1

        ax.plot(x, y, color = 'orange', alpha = 0.05)
        
    ax.set_title('Coercivity vs. B (log_mT)', fontsize=14)
    B_labels = [1, 10, 25, 50, 100, 300, 1000]
    ax.set_xticks(np.log10(B_labels))
    ax.set_xticklabels(B_labels)
    ax.set_xlabel('log B (mT)', fontsize=14)
    ax.set_ylabel('Normalized magnetization', fontsize=14)
    return ax
```

```python
def plot_results_2_components(example_data, trace_dataframe, n = 100):
    fig, ax = plt.subplots(figsize = (15,6))
    ax.scatter(example_data[0], example_data[1], color = 'grey', alpha = 0.6)
    x = np.linspace(0, example_data[0].max(), 100)
    def skew_Func(x, amp, mu, skew, sigma):
        return amp * np.exp(-0.5 * ((x - mu) / sigma)**2) * (1+real_erf((skew*x-mu)/sigma/np.sqrt(2)))
    
    for i in range(n):
        sample_row = trace_dataframe.sample(axis=0, n=1)
        y1 = skew_Func(x, amp =np.array(sample_row['amp1']),
                      mu = np.array(sample_row['mu1']), 
                      skew = np.array(sample_row['skew1']),
                      sigma = np.array(sample_row['sd1']))
        y2 = skew_Func(x, amp = np.array(sample_row['amp2']),
                      mu = np.array(sample_row['mu2']),
                      skew = np.array(sample_row['skew2']),
                      sigma = np.array(sample_row['sd2']))
        y = y1+y2
#         print(y1)
#         print(GaussFunc(x,1,1,1))
        ax.plot(x, y1, color = 'dodgerblue', alpha = 0.05)
        ax.plot(x, y2, color = 'coral', alpha = 0.05)
        ax.plot(x, y, color = 'orange', alpha = 0.05)
        
    ax.set_title('Coercivity vs. B (log_mT)', fontsize=14)
    B_labels = [1, 10, 25, 50, 100, 300, 1000]
    ax.set_xticks(np.log10(B_labels))
    ax.set_xticklabels(B_labels)
    ax.set_xlabel('log B (mT)', fontsize=14)
    ax.set_ylabel('Normalized magnetization', fontsize=14)
    return ax
```

```python
def plot_results_3_components(example_data, trace_dataframe, n = 100):
    fig, ax = plt.subplots(figsize = (15,6))
    ax.scatter(example_data[0], example_data[1], color = 'grey', alpha = 0.6)
    x = np.linspace(0, example_data[0].max(), 100)
    def skew_Func(x, amp, mu, skew, sigma):
        return amp * np.exp(-0.5 * ((x - mu) / sigma)**2) * (1+real_erf((skew*x-mu)/sigma/np.sqrt(2)))
    
    for i in range(n):
        sample_row = trace_dataframe.sample(axis=0, n=1)
        y1 = skew_Func(x, amp =np.array(sample_row['amp1']),
                      mu = np.array(sample_row['mu1']), 
                      skew = np.array(sample_row['skew1']),
                      sigma = np.array(sample_row['sd1']))
        y2 = skew_Func(x, amp = np.array(sample_row['amp2']),
                      mu = np.array(sample_row['mu2']),
                      skew = np.array(sample_row['skew2']),
                      sigma = np.array(sample_row['sd2']))
        y3 = skew_Func(x, amp = np.array(sample_row['amp3']),
                      mu = np.array(sample_row['mu3']),
                      skew = np.array(sample_row['skew3']),
                      sigma = np.array(sample_row['sd3']))
        
        y = y1+y2+y3
#         print(y1)
#         print(GaussFunc(x,1,1,1))
        ax.plot(x, y1, color = 'dodgerblue', alpha = 0.05)
        ax.plot(x, y2, color = 'coral', alpha = 0.05)
        ax.plot(x, y3, color = 'green', alpha = 0.05)
        ax.plot(x, y, color = 'orange', alpha = 0.05)
        
    ax.set_title('Coercivity vs. B (log_mT)', fontsize=14)
    B_labels = [1, 10, 25, 50, 100, 300, 1000]
    ax.set_xticks(np.log10(B_labels))
    ax.set_xticklabels(B_labels)
    ax.set_xlabel('log B (mT)', fontsize=14)
    ax.set_ylabel('Normalized magnetization', fontsize=14)
    return ax
```

```python
# Loading IRM data file
rockmag_data = pd.ExcelFile('../data/Rock_mag_data/SSRM2022C_IRMDB_export.xlsx')
VSM_bf_measurements = pd.read_excel(rockmag_data, 'VSM backfield measurements')
```

## DA4-r

```python
DA4_A_logBlist = get_VSM_bf_data(VSM_bf_measurements, 1, 0.05, 0.05)['Log_Blist'][2:]
DA4_A_dMdBlist = get_VSM_bf_data(VSM_bf_measurements, 1, 0.05, 0.05)['Derivative']
plt.figure(figsize=(10,5))
plt.scatter(DA4_A_logBlist, DA4_A_dMdBlist, c='gray', alpha=0.6, label='original data')

plt.xlabel('log B (mT)', fontsize=14)
plt.ylabel('dM/dB', fontsize=14)
plt.legend(fontsize=14)

plt.show()
```

## One component fitting

```python
with pm.Model() as model:  # model specifications in PyMC3 are wrapped in a with-statement
    # Define priors
    mu = pm.Uniform("mu", 0,3)
    sigma = pm.Uniform("sigma", 0,10)
    amp = pm.Uniform("amp", 0, 1)
    skew = pm.Uniform('skew', 0, 1)
#     shift = Uniform('shift', -10, 10)
#     precision = Uniform("std", 0, 1)
#     2 * norm.pdf(x) * norm.cdf(a*x)

    def skew_GaussFunc(x, amp, mu, skew, sigma):
        return amp * np.exp(-0.5 * ((x - mu) / sigma)**2) * (1+erf((skew*x-mu)/sigma/np.sqrt(2)))

#     def skew_GaussFunc(x, amp, mu, skew, shift, sigma):
#         return amp * np.exp(-0.5 * ((skew*(x-shift) - mu) / sigma)**2)
    
    # Define likelihood
    likelihood = pm.Normal("coercivity", 
                        mu=skew_GaussFunc(np.array(DA4_A_logBlist), amp, mu, skew, sigma), 
                        sd=0.01, 
                        observed=np.array(DA4_A_dMdBlist))

    # Inference!
    trace_1 = pm.sample(draws=20000, tune=20000, cores=4, chains=4, return_inferencedata=0)
```

```python
pm.save_trace(trace_1, 'code_output/Bayesian_unmixing/1_componenet_trace')
```

```python
trace_dataframe_1 = pm.trace_to_dataframe(trace_1)
trace_dataframe_1.head()
```

```python
ax_1_component = plot_results_1_component([DA4_A_logBlist, DA4_A_dMdBlist], trace_dataframe_1, 200)
```

## two component fitting

```python
with pm.Model() as model:
    amp1 = pm.Uniform('amp1', 0, 1.)
    mu1 = pm.Uniform('mu1', 0, 3.)
    sd1 = pm.Uniform('sd1', 0, 3.)
    skew1 = pm.Uniform('skew1', 0, 1.)
    
    amp2 = pm.Uniform('amp2', 0, 1.)
    mu2 = pm.Normal('mu2', 1.5, 2.5)
    sd2 = pm.Uniform('sd2', 0, 3.)
    skew2 = pm.Uniform('skew2', 0, 1.)
    
    
    def skew_GaussFunc(x, amp, mu, skew, sigma):
        return amp * np.exp(-0.5 * ((x - mu) / sigma)**2) * (1+erf((skew*x-mu)/sigma/np.sqrt(2)))
    
    likelihood = pm.Normal("coercivity", 
                        mu=skew_GaussFunc(DA4_A_logBlist, amp1, mu1, skew1, sd1)
                        +skew_GaussFunc(DA4_A_logBlist, amp2, mu2, skew2, sd2), 
                        sd=0.01, 
                        observed=DA4_A_dMdBlist)
    # Inference!

    trace_2 = pm.sample(draws=20000, tune=20000, cores=4, return_inferencedata=0, chains=4)
```

## three component fitting

```python
with pm.Model() as model:
    amp1 = pm.Uniform('amp1', 0, 1.)
    mu1 = pm.Uniform('mu1', 1.5, 2.0)
    sd1 = pm.Uniform('sd1', 0, 3.)
    skew1 = pm.Uniform('skew1', 0, 1.)
    
    amp2 = pm.Uniform('amp2', 0, 1.)
    mu2 = pm.Uniform('mu2', 0, 1.5)
    sd2 = pm.Uniform('sd2', 0, 3.)
    skew2 = pm.Uniform('skew2', 0, 1.)
    
    amp3 = pm.Uniform('amp3', 0, 1.)
    mu3 = pm.Uniform('mu3', 2., 3.)
    sd3 = pm.Uniform('sd3', 0, 3.)
    skew3 = pm.Uniform('skew3', 0, 1.)
    
    def skew_GaussFunc(x, amp, mu, skew, sigma):
        return amp * np.exp(-0.5 * ((x - mu) / sigma)**2) * (1+erf((skew*x-mu)/sigma/np.sqrt(2)))
    
    likelihood = pm.Normal("coercivity", 
                        mu=skew_GaussFunc(DA4_A_logBlist, amp1, mu1, skew1, sd1)
                        +skew_GaussFunc(DA4_A_logBlist, amp2, mu2, skew2, sd2)
                        +skew_GaussFunc(DA4_A_logBlist, amp3, mu3, skew3, sd3),
                        sd=0.01, 
                        observed=DA4_A_dMdBlist)
    # Inference!

    trace_3 = pm.sample(draws=20000, tune=20000, cores=4, return_inferencedata=0, chains=4)
```
