## Correct for holder value

```python
def func(x, a, b, c, d):
    return a + b*x + c*x**2 + d*x**3
```

```python
xdata = holder_data["T [ºC]"]
ydata = holder_data['k [m3]']
xdata
```

```python
plt.plot(xdata, ydata, 'b-', label='data')
```

```python
xdata = [item for item in xdata if not(pd.isnull(item)) == True]
ydata = [item for item in ydata if not(pd.isnull(item)) == True]
plt.plot(xdata, ydata, 'b-', label='data')
```

```python
popt, pcov = curve_fit(func, xdata, ydata)
```

```python
popt
```

```python
pcov
```

```python
fitted = [func(i,*popt) for i in xdata]
```

```python
fig,ax = plt.subplots()


ax.plot(xdata, ydata, 'pink', label='data')
ax.plot(xdata, fitted, 'b-', label=f'curve fit y={popt[0]}+{popt[1]}x +{popt[2]}x**2 + {popt[3]}x**3')
ax.set_xlabel('T [ºC]')
ax.set_ylabel(' Susceptibility [m3/kg]')
plt.legend()
```

```python
DA4_r_data["holder correction k [m3]"] = [func(i,*popt) for i in DA4_r_data["T [ºC]"]]
DA4_r_data["holder correction k [m3]"]
```

```python
DA4_r_data["k corrected for holder [m3]"] = DA4_r_data['k [m3]'] - DA4_r_data["holder correction k [m3]"]
DA4_r_data["k corrected for holder [m3]"]
```

```python
fig,ax = plt.subplots(figsize=(10,10))

ax.plot(DA4_r_data["T [ºC]"] , DA4_r_data["k corrected for holder [m3]"], label = 'DA4_r_data_corrected', color = 'black')



ax.set_xlabel('T [ºC]')
ax.set_ylabel(' Susceptibility [m3]')

ax.legend()
```
