# Notebook for displaying QDM maps of the Duluth anorthosite xenoliths vs. Duluth anorthosite

```python
import pmagpy.ipmag as ipmag
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
import imageio
from PIL import Image
import matplotlib.cm as cm
%config InlineBackend.figure_format = 'retina'
```

## DX1-4A NRM

```python
DX1_4A_Bz = loadmat('../data/QDM_Bz_maps/DX1_4A_Bz_uc0.mat')
DX1_4A_Bz = DX1_4A_Bz['Bz'].reshape(300,374)

DX1_4A_LED = Image.open('../data/QDM_Bz_maps/DX1_4A_LED.jpg').resize((374,300)).convert('LA')
```

```python
xticks_list = np.array([0, 75, 150, 225, 300, 375])
yticks_list = np.array([0, 75, 150, 225])
```

```python
vmin_QDM=-1e-5
vmax_QDM=1e-5

plt.figure(figsize = (37.4/3, 10))
# plt.imshow(DX1_4A_LED,alpha=0.6,cmap='Greys')
plt.imshow(np.flip(DX1_4A_Bz,0),cmap='seismic',vmin=vmin_QDM,vmax=vmax_QDM, alpha = .6)
cbar = plt.colorbar(fraction=0.05,shrink=0.85, format='%.e', ticks=np.linspace(vmin_QDM, vmax_QDM, 5))
cbar.ax.set_xlabel('Bz (T)', fontsize=16, labelpad=15)
cbar.ax.tick_params(labelsize=16)

cbar.update_ticks()
plt.xticks(ticks = xticks_list, labels = xticks_list* 4, fontsize=16)
plt.yticks(yticks_list,yticks_list* 4, fontsize=16)
plt.savefig('../figure/DX1_4A_Bz_plot_with_scale.png', bbox_inches='tight', dpi=300)
plt.show()
```
