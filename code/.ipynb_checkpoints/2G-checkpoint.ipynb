# Notebook for rewriting 2G file in IRM database to CIT format

```python
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd

import datetime
import pmagpy.ipmag as ipmag
import pmagpy.pmag as pmag
from IPython.display import display
%config InlineBackend.figure_format = 'retina'
```

## load IRMDB data

```python
# Loading IRM data file
specimen_2G = ['DA2-1d','DA4-1C', 'DA6-1c', 'DA7-2c', 'DX1-1b', 'DX1-4b', 'DX1-4c', 'DX1-5b']

rockmag_data = pd.ExcelFile('../data/Rock_mag_data/SSRM2022C_IRMDB_export.xlsx')
specimen_info = pd.read_excel(rockmag_data, 'Specimen Info')
specimen_info = specimen_info[1:]
specimen_info = specimen_info[specimen_info['Specimen_ID'].str.contains('|'.join(specimen_2G))].reset_index(drop=1)
remanence_data = pd.read_excel(rockmag_data, 'remanence measurements')
```

## rewrite files to CIT format

```python
for spec in specimen_2G:
    this_spec_az = specimen_info[specimen_info['Specimen_ID']==spec].Specimen_azimuth
    this_spec_pl = specimen_info[specimen_info['Specimen_ID']==spec].Specimen_plunge
    this_spec_remanence = remanence_data[remanence_data['specimen'] == spec][['specimen','Mx [Am2]','My [Am2]','Mz [Am2]','AF [mT]']].reset_index(drop=1)
    
    this_spec_remanence['My_corr'] = -this_spec_remanence['My [Am2]']
    this_spec_remanence['M_tot'] = np.sqrt(this_spec_remanence['Mx [Am2]']**2+
                                           this_spec_remanence['My_corr']**2+
                                           this_spec_remanence['Mz [Am2]']**2)
    this_spec_remanence['spec_dec'] = pmag.cart2dir(this_spec_remanence[['Mx [Am2]','My_corr','Mz [Am2]']].to_numpy()).T[0]
    this_spec_remanence['spec_inc'] = pmag.cart2dir(this_spec_remanence[['Mx [Am2]','My_corr','Mz [Am2]']].to_numpy()).T[1]
    
    this_spec_remanence['geo_dec'] = [pmag.dogeo(this_spec_remanence['spec_dec'].iloc[i], 
                                             this_spec_remanence['spec_inc'].iloc[i], 
                                                 this_spec_az,  
                                                 -(90-this_spec_pl))[0] for i in range(len(this_spec_remanence['spec_dec']))]
    this_spec_remanence['geo_inc'] = [pmag.dogeo(this_spec_remanence['spec_dec'].iloc[i], 
                                                 this_spec_remanence['spec_inc'].iloc[i], 
                                                 this_spec_az,  
                                                 -(90-this_spec_pl))[1] for i in range(len(this_spec_remanence['spec_dec']))]
    
    # write CIT format file
    this_spec_lines = []
    header_line = spec+'\n'
#     print('{:>5}'.format((this_spec_az.values[0]+90)))
    orientation_line =  '{:>7}'.format(0)+ ' ' + '{:>5}'.format((this_spec_az.values[0]+90)) \
    + ' ' +'{:>5}'.format(90-this_spec_pl.values[0]) \
    + ' ' +'{:>5}'.format(0.0)+ ' ' +'{:>5}'.format(0.0)+' '+'{:>5}'.format(1.0)+'\n'
    this_spec_lines.append(header_line)
    this_spec_lines.append(orientation_line)
#     print(orientation_line)
    for i in range(this_spec_remanence.shape[0]):
        if int(this_spec_remanence['AF [mT]'].iloc[i]*10) == 0:
            this_line = 'NRM   '\
            +' '+'{:05.1f}'.format(this_spec_remanence['geo_dec'].iloc[i]) \
            +' '+'{:05.1f}'.format(this_spec_remanence['geo_inc'].iloc[i]) \
            +' '+'{:05.1f}'.format(this_spec_remanence['geo_dec'].iloc[i]) \
            +' '+'{:05.1f}'.format(this_spec_remanence['geo_inc'].iloc[i]) \
            +' '+'{:05.2e}'.format(this_spec_remanence['M_tot'].iloc[i]*1e3) \
            +' '+'000.1' \
            +' '+'{:05.1f}'.format(this_spec_remanence['spec_dec'].iloc[i]) \
            +' '+'{:05.1f}'.format(this_spec_remanence['spec_inc'].iloc[i]) \
            +' '+'1.000000'+' '+'1.000000'+' '+'1.000000'+' '+'IRM_2G' \
            +' '+'2022-06-10 12:00:00\n'
            this_spec_lines.append(this_line)
            print(this_line)
        else:
            this_line = 'AF'+ '{:>4}'.format(int(this_spec_remanence['AF [mT]'].iloc[i]*10)) \
            +' '+'{:05.1f}'.format(this_spec_remanence['geo_dec'].iloc[i]) \
            +' '+'{:05.1f}'.format(this_spec_remanence['geo_inc'].iloc[i]) \
            +' '+'{:05.1f}'.format(this_spec_remanence['geo_dec'].iloc[i]) \
            +' '+'{:05.1f}'.format(this_spec_remanence['geo_inc'].iloc[i]) \
            +' '+'{:05.2e}'.format(this_spec_remanence['M_tot'].iloc[i]*1e3) \
            +' '+'000.1' \
            +' '+'{:05.1f}'.format(this_spec_remanence['spec_dec'].iloc[i]) \
            +' '+'{:05.1f}'.format(this_spec_remanence['spec_inc'].iloc[i]) \
            +' '+'1.000000'+' '+'1.000000'+' '+'1.000000'+' '+'IRM_2G' \
            +' '+'2022-06-10 12:00:00\n'
            this_spec_lines.append(this_line)
            print(this_line)
#     this_spec_file = open('../data/2G_data/'+spec, 'w')
#     this_spec_file.writelines(this_spec_lines)
    
    this_spec_file.close()
#     display(this_spec_remanence)
    
```

## plot MagIC data and compare with known data

```python
location_file = pd.read_csv('../data/2G_data/DADX_locations.txt', sep='\t', header=1)
location_file
```

```python
MagIC_spec_file = pd.read_csv('../data/2G_data/specimens.txt', sep='\t', header=1)
MagIC_spec_file['vgp_lon'] = [pmag.dia_vgp(MagIC_spec_file['dir_dec'][i], MagIC_spec_file['dir_inc'][i], 0, 
                                           location_file['lat_n'][0], location_file['lon_e'][0])[0] for i in range(MagIC_spec_file.shape[0])]
MagIC_spec_file['vgp_lat'] = [pmag.dia_vgp(MagIC_spec_file['dir_dec'][i], MagIC_spec_file['dir_inc'][i], 0, 
                                           location_file['lat_n'][0], location_file['lon_e'][0])[1] for i in range(MagIC_spec_file.shape[0])]

mc_geo = MagIC_spec_file[(MagIC_spec_file['dir_tilt_correction'] == 0) & 
                                (MagIC_spec_file['dir_comp'] == 'mc')]
mc_geo = mc_geo[mc_geo['site'] != 'DA2'].reset_index(drop=1)
mc_geo_DA = mc_geo[mc_geo['site'].str.contains('DA')].reset_index(drop=1)
mc_geo_DX = mc_geo[mc_geo['site'].str.contains('DX')].reset_index(drop=1)

mc_geo_mean = ipmag.fisher_mean(mc_geo['dir_dec'], mc_geo['dir_inc'])
mc_geo_mean_pole = ipmag.fisher_mean(mc_geo['vgp_lon'], mc_geo['vgp_lat'])

# dir_geo_coord
```

```python
plt.figure(figsize=(6,6))
ipmag.plot_net()
ipmag.plot_di(mc_geo_DA['dir_dec'], mc_geo_DA['dir_inc'], color='r', alpha=0.6, marker='d', markersize=50, label='mid-coercivity component\nDA')
ipmag.plot_di(mc_geo_DX['dir_dec'], mc_geo_DX['dir_inc'], color='r', alpha=0.6, marker='s',markersize=50, label='mid-coercivity component\nDX')

ipmag.plot_di_mean(mc_geo_mean['dec'], mc_geo_mean['inc'], mc_geo_mean['alpha95'], markersize=30, color='r', marker='o')

ipmag.plot_di_mean(292.6, 54.8, 7.5,
              label='Anorthositic series (Beck, 1970)',markersize=30, color='k')

plt.legend(fontsize=16, loc=8)
# plt.savefig('code_output/2G_equal_area.png',dpi=300, bbox_inches='tight')
plt.show()
```

```python
two_Euler_pole_lon = np.loadtxt('../Data/mcplates_results/pole_lon_positions_2_true.txt')
two_Euler_pole_lat = np.loadtxt('../Data/mcplates_results/pole_lat_positions_2_true.txt')

m = ipmag.make_orthographic_map(central_latitude=35,central_longitude=200)
plt.figure(num=1,figsize=(10, 10))

fisher0 = ipmag.fisher_mean(two_Euler_pole_lon[:,0]+180,two_Euler_pole_lat[:,0])
theta0 = ipmag.fisher_angular_deviation(two_Euler_pole_lon[:,0]+180,two_Euler_pole_lat[:,0])
fisher0['theta'] = theta0
ipmag.plot_pole(m,fisher0['dec'],fisher0['inc'],theta0,color='C0',label='1110 Ma')

fisher1 = ipmag.fisher_mean(two_Euler_pole_lon[:,1]+180,two_Euler_pole_lat[:,1])
theta1 = ipmag.fisher_angular_deviation(two_Euler_pole_lon[:,1]+180,two_Euler_pole_lat[:,1])
fisher1['theta'] = theta1
ipmag.plot_pole(m,fisher1['dec'],fisher1['inc'],theta1,color='C1',label='1105 Ma')

fisher2 = ipmag.fisher_mean(two_Euler_pole_lon[:,2]+180,two_Euler_pole_lat[:,2])
theta2 = ipmag.fisher_angular_deviation(two_Euler_pole_lon[:,2]+180,two_Euler_pole_lat[:,2])
fisher2['theta'] = theta2
ipmag.plot_pole(m,fisher2['dec'],fisher2['inc'],theta2,color='C2',label='1100 Ma')

fisher3 = ipmag.fisher_mean(two_Euler_pole_lon[:,3]+180,two_Euler_pole_lat[:,3])
theta3 = ipmag.fisher_angular_deviation(two_Euler_pole_lon[:,3]+180,two_Euler_pole_lat[:,3])
fisher3['theta'] = theta3
ipmag.plot_pole(m,fisher3['dec'],fisher3['inc'],theta3,color='C3',label='1095 Ma')

fisher4 = ipmag.fisher_mean(two_Euler_pole_lon[:,4]+180,two_Euler_pole_lat[:,4])
theta4 = ipmag.fisher_angular_deviation(two_Euler_pole_lon[:,4]+180,two_Euler_pole_lat[:,4])
fisher4['theta'] = theta4
ipmag.plot_pole(m,fisher4['dec'],fisher4['inc'],theta4,color='C4',label='1090 Ma')

fisher5 = ipmag.fisher_mean(two_Euler_pole_lon[:,5]+180,two_Euler_pole_lat[:,5])
theta5 = ipmag.fisher_angular_deviation(two_Euler_pole_lon[:,5]+180,two_Euler_pole_lat[:,5])
fisher5['theta'] = theta5
ipmag.plot_pole(m,fisher5['dec'],fisher5['inc'],theta5,color='C5',label='1085 Ma')

fisher6 = ipmag.fisher_mean(two_Euler_pole_lon[:,6]+180,two_Euler_pole_lat[:,6])
theta6 = ipmag.fisher_angular_deviation(two_Euler_pole_lon[:,6]+180,two_Euler_pole_lat[:,6])
fisher6['theta'] = theta6
ipmag.plot_pole(m,fisher6['dec'],fisher6['inc'],theta6,color='C6',label='1080 Ma')

ipmag.plot_vgp(m, mc_geo_DA['vgp_lon'].tolist(), mc_geo_DA['vgp_lat'].tolist(), color='lightcoral', marker='d', markersize=50, label='DA vgp')
ipmag.plot_vgp(m, mc_geo_DX['vgp_lon'].tolist(), mc_geo_DX['vgp_lat'].tolist(), color='lightcoral', marker='s', markersize=50, label='DX vgp')

ipmag.plot_pole(m, mc_geo_mean_pole['dec'], mc_geo_mean_pole['inc'], mc_geo_mean_pole['alpha95'], color='lightcoral', marker='o', markersize=100)

plt.legend(fontsize=12, loc=4)
plt.savefig('code_output/2G_vgp.png', dpi=300, bbox_inches='tight')
plt.show()
```
